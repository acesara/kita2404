```python
titles = soup.find_all(class_='Title_title__s9o0D')

title_list = []
for idx, title in enumerate(titles[:100]):
    title_list.append(title.text)
```

```python
prices = soup.find_all(class_='Price_price__GqXqo')
price_list = []
for idx, price in enumerate(prices[:100]):
    price_list.append(price.text[:-2])
```

```python
types = soup.find_all(class_='Price_text__pRk_f')
type_list = []
for idx, type in enumerate(types[:100]):
    type_list.append(type.text.strip())
```

```python
data = {
    'TITLE': title_list,
    'PRICE': price_list,
    'TYPE': type_list
}
df = pd.DataFrame(data)
print(df)
```

```python
# 개발자 도구 copy.selector

import re
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"
    }
url = 'https://news.daum.net/politics#1'
r = rq.get(url, headers=headers)
html = r.text
soup = BeautifulSoup(html, 'html.parser')

links = soup.select('a.link_txt')

for i, tag in enumerate(links, start=1):
    text = tag.get_text().strip().replace(',','')
    matches = re.findall('[가-힣0-9]+', text)
    print(f"{i}. {' '.join(matches)}")
```
