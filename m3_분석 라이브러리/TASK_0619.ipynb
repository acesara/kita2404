Task1_0619. 다음 사항을 수행하세요.
- 첫 번째로 매칭되는 'p' 태그 찾기
- 모든 'p' 태그 찾기
- 클래스가 'content'인 첫 번째 'p' 태그 찾기
- 클래스가 'content'인 모든 'p' 태그 찾기
- 특정 'p' 태그의 모든 부모 태그 찾기
- 특정 'p' 태그의 첫 번째 부모 태그 찾기
- 특정 'p' 태그의 다음 형제 태그 찾기
- 특정 'p' 태그의 이전 형제 태그 찾기
- 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
- 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 

```python
from bs4 import BeautifulSoup

html_content = '<html><body><h1>Title</h1><p class="content">First paragraph.</p><p class="content">Second paragraph.</p></body></html>'
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())
```

```python
# Task1_0619. 다음 사항을 수행하세요.
# 첫 번째로 매칭되는 'p' 태그 찾기
# 모든 'p' 태그 찾기
# 클래스가 'content'인 첫 번째 'p' 태그 찾기
# 클래스가 'content'인 모든 'p' 태그 찾기
# 특정 'p' 태그의 모든 부모 태그 찾기
# 특정 'p' 태그의 첫 번째 부모 태그 찾기
# 특정 'p' 태그의 다음 형제 태그 찾기
# 특정 'p' 태그의 이전 형제 태그 찾기
# 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기
# 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 
```

```python
# 첫 번째로 매칭되는 'p' 태그 찾기
p_tag = soup.find('p')
print(p_tag.text)
```

```python
first_p = soup.find('p')
print(first_p.get_text())
```

```python
# 모든 'p' 태그 찾기

all_p_tags = soup.find_all('p')
for tag in all_p_tags:
    print(tag.text)
```

```python
all_p = soup.find_all('p')
for a in all_p:
    print(a.text)
```

```python
# 클래스가 'content'인 첫 번째 'p' 태그 찾기

first_tag = soup.select_one('p')
print(first_tag)
```

```python
first_content_p = soup.select_one('p.content')
print(first_content_p.get_text())
```

```python
# 클래스가 'content'인 모든 'p' 태그 찾기

all_p_tags = soup.select('p')
for tag in all_p_tags:
    print(all_p_tags) 
```

```python
all_content_p = soup.select('p.content')
for p in all_content_p:
    print(p.get_text())
```

```python
specific_p = soup.find('p', class_='content')
specific_p
```

```python
# 특정 'p' 태그의 모든 부모 태그 찾기

print(p_tag.find_parents())
```

```python
# 특정 'p' 태그의 모든 부모 태그 찾기
# [document] -> <html> -> <body> -> <p> : 최상위 요소 document에서 시작하여 p 태그까지의 모든 부모 요소를 출력
specific_p = soup.find('p', class_='content')
print(specific_p,'\n')

parents = specific_p.find_parents()
#print(parents)
for parent in parents:
   print(parent.name)
```

```python
# 특정 'p' 태그의 첫 번째 부모 태그 찾기

parent = specific_p.find_parent()
print(parent.name)
```

```python
# 특정 'p' 태그의 다음 형제 태그 찾기

print(p_tag.find_next_sibling())
```

```python
next_sibling = specific_p.find_next_sibling()
print(next_sibling.get_text())
```

```python
# 특정 'p' 태그의 이전 형제 태그 찾기

print(p_tag.find_previous_sibling())
```

```python
previous_sibling = specific_p.find_previous_sibling()
print(previous_sibling)
```

```python
# 특정 'p' 태그 다음에 위치한 모든 태그나 문자열 찾기

next_elements = specific_p.find_next()
print(next_elements)
```

```python
# 특정 'p' 태그 이전에 위치한 모든 태그나 문자열 찾기 

previous_elements = specific_p.find_previous()
print(previous_elements)
```

Task2_0619. ID를 이용해서 '네이버 뉴스' 추출하세요.

```python
import requests
from bs4 import BeautifulSoup

# 예시 네이버 뉴스 URL

url = 'https://news.naver.com/'

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)
response
soup = BeautifulSoup(response.content, 'html.parser')

print(soup.prettify())
```

```python
link_tag = soup.select_one('#browserTitleArea')
print(link_tag.get_text())
```

```python
link_tag = soup.select_one('title#browserTitleArea')
link_tag.get_text()
```

```python
title_elements = soup.find('title', id='browserTitleArea')
print(title_elements)
title_elements.get_text()
```

Task3_0619. soup.find_all(class_='Nitem_link_menu') 대신에 select를 이용하여 동일한 결과를 출력하세요.

```python
soup = BeautifulSoup(response.content, 'html.parser')
links = soup.find_all(class_='Nitem_link_menu')
print(links)
for link in links:
    print(link.text)
```

```python
tags = soup.select('.Nitem_link_menu')
for tag in tags:
    print(tag.text)
```

```python
cats = soup.select('.Nitem_link_menu')
for idx, cat in enumerate(cats):
    print(f"{idx + 1}: {cat.get_text().strip()}")
```
